```python
def peredelka(mat):
    #вычисление среднего
    sredn=[]
    for i in range(n):
        for j in range(n):
            if not isnan(mat[i][j]):
                sredn.append(mat[i][j])
    s=np.mean(sredn)
    print('')
    print('среднее значение элементов исходной матрицы',s.round(5))
    print('')
    
    #если массив пустой
    if isnan(s):
        mat=np.zeros((n,n))

    #замена NAN средним значением через маску
    mask=np.empty((n,n),dtype=bool)
    for i in range(n):
        for j in range(n):
            mask[i][j]=isnan(mat[i][j])
    mat[mask]=s.round(4)
    print('матрица после замены')
    print(mat)
```

```python
n=6 # размер матрицы
mat=np.random.random(size=(n, n))  # создание матрицы
mat=mat.round(3)
for i in range(10):     #  заменяю 10 случайных элементов на NAN
    mat[np.random.randint(0,n)][np.random.randint(0,n)]=np.nan
print('исходная матрица')
print(mat)
peredelka(mat)
```

```python
def norm_mass(mass):
    m=np.max(mass)
    for i in range(len(mass)):
        mass[i]=(mass[i]/m).round(4)
    print('нормализованный массив')
    print(mass)
```

```python
# Напишите функцию, которая возвращает  𝑛×𝑚  
# целочисленную матрицу с заданным значением на границах матрицы и с заданным значением внутри
```

```python
def grani (n,m,gr,ins):
    mat=np.zeros(shape=(n,m), dtype=int) 
    mat[:,0]=gr
    mat[:,-1]=gr
    mat[:,1:-1]=ins
    print('матрица размерностью',n,'на',m)
    print('на границах',gr,'внутри',ins)
    print(mat)
```

```python
# Напишите функцию, которая сортирует строки данной матрицы по значениям заданного столбца в порядке убывания.
```

```python
n=random.randint(4,20) # размер матрицы - рандом
mat=np.random.randint(10, 60, size=(n, n)) # матрица с рандом-значениями
print('исходная матрица')
print(mat)
```

```python
def sortm(st):
    for k in range(n-1):
        for i in range(n-1):
            if mat[i,st]<mat[i+1,st]:
                mat[[i,i+1]]=mat[[i+1,i]]
    print('матрица после сортировки по',st,'столбцу')
    print(mat)
```

```python
st=6 # номер столбца для сортировки, первый столбец - "0"
if st>(n-1):
    print('в матрице нет такого столбца')
else:
    sortm(st)
```
